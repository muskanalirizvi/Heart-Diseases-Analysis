How many people have heartdisease? and How many don't have heartdisease in the Dataset?

```python
data.columns
```

```python
target_counts = data['target'].value_counts()
target_counts
```

```python

target_counts.plot(kind='bar')
plt.xlabel('target')
plt.ylabel('Count')
plt.title('Count of Target Values')
plt.xticks([0,1],['Low risk', 'High risk'])
plt.show()
```

Which Sex Has The Most Heart Disease [ Male or Female ]?

```python
sex_counts = data['sex'].value_counts()
sex_counts
```

```python
sex_counts.plot(kind = 'bar')
plt.xlabel('sex')
plt.ylabel('Count')
plt.title('Count of sex with more most heartdisease')
plt.xticks([0,1], ['Male', 'Female'])
plt.show()
```

Find Gender Distribution According to The Target Variable.

```python
sns.countplot(x= 'sex', hue= 'target', data= data)
plt.xticks([1,0], ['Male', 'Female'])
plt.legend(labels = ['No_disease', 'Disease'])
plt.show()
```

Check Age Distribution In The Dataset

 Check Chest Pain Type
chest pain type (4 values)
Value 0: typical angina
Value 1: atypical angina
Value 2: non-anginal pain
Value 3: asymptomatic

```python
cp_counts = data['cp'].value_counts()
cp_counts
```

```python
cp_counts.plot(kind = 'bar')
plt.xlabel('cp')
plt.ylabel('Count')
plt.title('Check chest pain types')
plt.xticks([0,1,2,3],["typical angina","non-anginal pain","atypical angina","asymptomatic "])
plt.show()
```

Show The Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

Show Fasting Blood Sugar Distribution According To Target Variable

```python
sns.countplot(x='fbs',hue='target',data=data)
plt.legend(labels=['No-Disease','Diseases'])
plt.show()
```

Compare Resting Blood Pressure As Per Sex Column

```python
g = sns.FacetGrid(data,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend(labels=['Male','Female'])
plt.show()
```

Plot Continuous Variables

```python
cate_val = []
cont_val = []
for column in data.columns:
    if len(data[column].unique())<=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cont_val
```

```python
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
